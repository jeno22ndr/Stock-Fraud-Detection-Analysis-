```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
```

```python
df = pd.read_csv('/content/Stock Market Anomaly Detection Dataset.csv')
```

**Reasoning**:
Display the first 5 rows of the DataFrame, a concise summary, and descriptive statistics to understand the data. Check for missing values to identify any data quality issues.



```python
display(df.head())
display(df.info())
display(df.describe())
display(df.isnull().sum())
```

**Reasoning**:
Select numerical columns for scaling, excluding 'Sr.no', instantiate and fit a StandardScaler, transform the data, and convert the scaled array back to a DataFrame with original column names.



```python
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
numerical_cols.remove('Sr.no')

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[numerical_cols])
scaled_df = pd.DataFrame(scaled_features, columns=numerical_cols)
```

```python
# 1. Instantiate IsolationForest
isolation_forest = IsolationForest(contamination='auto', random_state=42)

# 2. Build Autoencoder model
input_dim = scaled_df.shape[1]
encoding_dim = int(input_dim / 2) # Example: reduce dimension by half

input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation="relu")(input_layer)
decoder = Dense(input_dim, activation="linear")(encoder)

autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

print("Isolation Forest model instantiated.")
autoencoder.summary()
```

```python
# Fit Isolation Forest
isolation_forest.fit(scaled_df)

# Train Autoencoder
history = autoencoder.fit(scaled_df, scaled_df,
                          epochs=50,
                          batch_size=32,
                          shuffle=True,
                          validation_split=0.2,
                          verbose=0)

print("Isolation Forest model fitted.")
print("Autoencoder model trained.")
```

```python
# 1. Predict anomaly scores using Isolation Forest
df['isolation_forest_score'] = isolation_forest.decision_function(scaled_df)

# 2. Predict reconstructed data using Autoencoder
reconstructed_data = autoencoder.predict(scaled_df)

# 3. Calculate reconstruction error for Autoencoder (Mean Squared Error)
mse = np.mean(np.power(scaled_df - reconstructed_data, 2), axis=1)
df['autoencoder_error'] = mse

# 4. Display the first few rows of the updated dataframe
display(df.head())
```

```python
# 1. Define threshold for Isolation Forest scores (using 5th percentile)
isolation_forest_threshold = df['isolation_forest_score'].quantile(0.05)
print(f'Isolation Forest Threshold (5th percentile): {isolation_forest_threshold}')

# 2. Create boolean column for Isolation Forest anomalies
df['is_anomaly_isolation_forest'] = df['isolation_forest_score'] <= isolation_forest_threshold

# 3. Define threshold for Autoencoder reconstruction errors (using 95th percentile)
autoencoder_threshold = df['autoencoder_error'].quantile(0.95)
print(f'Autoencoder Threshold (95th percentile): {autoencoder_threshold}')

# 4. Create boolean column for Autoencoder anomalies
df['is_anomaly_autoencoder'] = df['autoencoder_error'] >= autoencoder_threshold

# 5. Optionally, create a combined anomaly column
df['is_anomaly_combined'] = df['is_anomaly_isolation_forest'] | df['is_anomaly_autoencoder']

# 6. Display the number of anomalies identified by each model
print("\nNumber of Anomalies Identified:")
print(f"Isolation Forest: {df['is_anomaly_isolation_forest'].sum()}")
print(f"Autoencoder: {df['is_anomaly_autoencoder'].sum()}")
print(f"Combined: {df['is_anomaly_combined'].sum()}")

display(df.head())
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['isolation_forest_score'], df['autoencoder_error'], c=df['is_anomaly_combined'], cmap='coolwarm', alpha=0.6)
plt.xlabel('Isolation Forest Anomaly Score')
plt.ylabel('Autoencoder Reconstruction Error')
plt.title('Anomaly Detection Results: Isolation Forest vs. Autoencoder')

# Create legend manually
legend_labels = {0: 'Normal', 1: 'Anomaly'}
colors = ['#440154ff', '#fde725ff'] # Example colors from coolwarm cmap

legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=label,
                              markerfacecolor=colors[i], markersize=10)
                   for i, label in legend_labels.items()]

plt.legend(handles=legend_elements, title='Anomaly Status')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for Isolation Forest
param_grid_iso = {
    'n_estimators': [100, 200, 300],
    'max_samples': ['auto', 0.8, 0.9],
    'contamination': ['auto', 0.01, 0.05],
    'max_features': [1.0, 0.8, 0.9]
}

# Instantiate Isolation Forest
iso_forest = IsolationForest(random_state=42)

# Instantiate GridSearchCV
grid_search_iso = GridSearchCV(iso_forest, param_grid_iso, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)

# Fit GridSearchCV
# We will use the scaled data for tuning
grid_search_iso.fit(scaled_df)

# Get the best parameters and best score
best_params_iso = grid_search_iso.best_params_
best_score_iso = grid_search_iso.best_score_

print(f"Best parameters for Isolation Forest: {best_params_iso}")
print(f"Best negative MSE score for Isolation Forest: {best_score_iso}")

# Update the Isolation Forest model with the best parameters
isolation_forest = IsolationForest(**best_params_iso, random_state=42)
isolation_forest.fit(scaled_df)
```

**Reasoning**:
Tune the Autoencoder model by training it with different combinations of epochs and batch sizes, evaluating the loss for each combination. We will keep track of the training history to observe the loss over epochs.

```python
# Define parameters for Autoencoder tuning
epochs_list = [50, 100]
batch_size_list = [32, 64]

best_loss = float('inf')
best_epochs = None
best_batch_size = None
autoencoder_history = {}

# Tune Autoencoder
for epochs in epochs_list:
    for batch_size in batch_size_list:
        print(f"Training Autoencoder with epochs={epochs}, batch_size={batch_size}")

        # Rebuild the Autoencoder model to reset weights
        input_dim = scaled_df.shape[1]
        encoding_dim = int(input_dim / 2) # Example: reduce dimension by half

        input_layer = Input(shape=(input_dim,))
        encoder = Dense(encoding_dim, activation="relu")(input_layer)
        decoder = Dense(input_dim, activation="linear")(encoder)

        autoencoder = Model(inputs=input_layer, outputs=decoder)
        autoencoder.compile(optimizer='adam', loss='mean_squared_error')

        # Train Autoencoder
        history = autoencoder.fit(scaled_df, scaled_df,
                                  epochs=epochs,
                                  batch_size=batch_size,
                                  shuffle=True,
                                  validation_split=0.2,
                                  verbose=0)

        # Evaluate the model on the validation set
        # Note: For anomaly detection, we are primarily interested in the reconstruction error,
        # which is the loss function we've used.
        validation_loss = history.history['val_loss'][-1]
        print(f"Validation Loss: {validation_loss}")

        # Store history
        autoencoder_history[f'epochs_{epochs}_batch_{batch_size}'] = history.history

        # Check if this combination is the best
        if validation_loss < best_loss:
            best_loss = validation_loss
            best_epochs = epochs
            best_batch_size = batch_size

print(f"\nBest parameters for Autoencoder: Epochs={best_epochs}, Batch Size={best_batch_size}")
print(f"Best validation loss for Autoencoder: {best_loss}")

# Retrain the Autoencoder with the best parameters
print("\nRetraining Autoencoder with best parameters...")
input_dim = scaled_df.shape[1]
encoding_dim = int(input_dim / 2)

input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation="relu")(input_layer)
decoder = Dense(input_dim, activation="linear")(encoder)

autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

history = autoencoder.fit(scaled_df, scaled_df,
                          epochs=best_epochs,
                          batch_size=best_batch_size,
                          shuffle=True,
                          validation_split=0.2,
                          verbose=0)

print("Autoencoder model retrained with best parameters.")
```
